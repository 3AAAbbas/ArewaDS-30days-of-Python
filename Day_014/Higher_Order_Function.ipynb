### 1. Explain the difference between map, filter, and reduce.


Map, filter, and reduce are three higher-order functions commonly used in functional programming and some modern programming languages. They are often associated with list (or iterable) operations and can be used to process data in a concise and expressive manner.

### i. Map Function:

- *Purpose:* map applies a given function to all items in an iterable (e.g., a list) and returns a new iterable containing the results.

- *Syntax* (in Python): map(function, iterable)

Example (in Python):



```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(squared_numbers)
```

### ii. Filter Function:

*Purpose:* filter constructs a new iterable from elements of the original iterable for which a given function returns True.

*Syntax* (in Python): filter(function, iterable)

Example (in Python):


```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(even_numbers)
```

### iii. reduce Function:

*Purpose:* reduce takes a binary function and a sequence and applies the function cumulatively to the elements of the sequence, reducing it to a single value.

*Syntax (in Python):* functools.reduce(function, iterable[, initializer])

Example (in Python):


```python
from functools import reduce
numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
```

### 2. Explain the difference between higher order function, closure and decorator


### i. Higher-order Function:

A higher-order function is a function that can take one or more functions as arguments or can return a function as its result.

Example (in Python):


```python
def apply_twice(func, arg):
    return func(func(arg))

def square(x):
    return x * x

result = apply_twice(square, 2)
```

### ii. Closure:
 A closure is a function object that has access to variables in its lexical scope, even when the function is called outside that scope.

Example (in Python):


```python
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

closure_instance = outer_function(10)
result = closure_instance(5)
```

### iii. Decorator:
A decorator is a design pattern in Python that allows the modification or extension of a function or class. It is implemented using functions (or classes) that take a function as input and return a new function that usually extends or modifies the behavior of the original function.

Example (in Python):


```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
